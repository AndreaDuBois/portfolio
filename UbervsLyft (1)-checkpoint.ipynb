```python
df['Distance'] = '0-1'
# df['Distance'] = np.where((df['distance'] > 0) & (df['distance'] <1),'0-1', df['Distance'])
df['Distance'] = np.where((df['distance'] > 1) & (df['distance'] <2),'1-2', df['Distance'])
df['Distance'] = np.where((df['distance'] > 2) & (df['distance'] <3),'2-3', df['Distance'])
df['Distance'] = np.where((df['distance'] > 3) & (df['distance'] <4),'3-4', df['Distance'])
df['Distance'] = np.where((df['distance'] > 4) & (df['distance'] <5),'4-5', df['Distance'])
df['Distance'] = np.where((df['distance'] > 5) & (df['distance'] <6),'5-6', df['Distance'])
df['Distance'] = np.where((df['distance'] > 6) & (df['distance'] <7),'6-7', df['Distance'])
df['Distance'] = np.where((df['distance'] > 7) & (df['distance'] <8),'7-8', df['Distance'])


```

```python
df.Distance.value_counts()
```

```python
df = df[df["Distance"]<'7-8']
df.sort_values('Distance',ascending = True, inplace = True) #sort descending with highest % of wins
sns.barplot(data=df, x="Distance", y="price", hue="cab_type")
```
